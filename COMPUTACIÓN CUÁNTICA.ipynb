## Crear un sistema cuntico complejos y sacar sus probabilidades

```python
from random import randrange, randint
from math import sqrt
def random_complex_quantum_state(epsilon=0):
    
    a = randint(-100, 100)/100
    b = randint(-100, 100)/100
    c = randint(-100, 100)/100
    d = randint(-100, 100)/100
    
    
    #a , b = float(input('Ingrese a: ')), float(input('Ingrese b '))
    #c , d = float(input('Ingrese c: ')), float(input('Ingrese d '))
    
    z1 = complex(a,b)
    z2 = complex(c,b)
    
    quantum_state = [ [z1], [z2] ]
    nz1 = abs(z1) #norma de z1
    nz2 = abs(z2) #norma de z2
    norma = sqrt(z1.real**2 + z1.imag**2 + z2.real**2 + z2.imag**2)
    print("La norma original es: ", norma)
    
    if abs(1-norma) <= epsilon:
        print("Ingresaste un estado cuntico vlido.")
        quantum_state = [ [z1],[z2] ]
        proba_0 = abs(z1)**2
        proba_1 = abs(z2)**2

    else:
        print("Esto no representa un estado cuntico vlido, pero ahorita te lo hago...")
        norma_n = sqrt((nz1/norma)**2 + (nz2/norma)**2) #norma del nuevo vector
        nz1n = complex(z1.real/norma,z1.imag/norma)
        nz2n = complex(z2.real/norma,z2.imag/norma)
        print("La nueva norma es: ", norma_n)
        quantum_state = [ [nz1n], [nz2n] ] #vector normalizado 
        
        proba_0 = abs(nz1n)**2
        proba_1 = abs(nz2n)**2
        
    print("Recuerde que la primera entrada del vector corresponde a |0> y la segunda entrada corresponde a |1>.")
    print("La primera entrada del vector corresponde a |0> con probabilidad = ", proba_0)
    print("La primera entrada del vector corresponde a |1> con probabilidad = ", proba_1)
    print("La suma de las probabilidades es: ", proba_0 + proba_1)
    
    return quantum_state
```

## Imprimir Matriz de Nuevo

```python
def printMatrix(m, literal='M'):
    from IPython.display import display, Math
    
    """
    Recibe una matriz m y una literal para que se muestre la asignacin
    Ej:
        M = [1 0 3]
    """
    rows = ""
    
    if type(m[0]) == list:
        for row in [[str(e) for e in r] for r in m]:
            rows += ' &'.join(row)
            rows += r'\\'
    else: # Si es un vector fila
        for row in [str(e) for e in m]:
            rows += f'{row} &'
        rows = rows[:-1]
        
    st = f'{literal} = \mymatrix%{"r"*len(m)}$%{rows}$'.replace('%', '{').replace('$', '}')
    display(Math(st))
```

## Tranpuesta de una Matriz

```python
def TransMatrix(M=None):
    if not M:
        f = int(input("Ingresa el # filas: "))
        c = int(input("Ingresa el # de columnas: "))
        print("Captura de elementos: ")
        M = []

        #llenando de ceros M lrptm
        for i in range(f):
            M.append([])
            for j in range(c):
                M[i].append(0)

        for i in range(f):
            for j in range(c):
                M[i][j] = int(input(f'Ingresa el elemento {i}, {j}: '))
                #M.append(M[i][j])

        print('La matriz es: ', M)
    else:
        f = len(M)
        c = len(M[0])
        
    T = []
        
    #llenando de ceros T lrptm
    for i in range(c):
        T.append([])
        for j in range(f):
            T[i].append(0)

    for i in range(c):
        for j in range(f):
            T[i][j] = M[j][i]
            #T.append(T[i][j])

    print('La transpuesta es: ', T)

    return M, T
```

## Producto Tensorial

```python
def tensorProduct(A=None, B=None, prints=True):
    def getMatrix():
        f = int(input("Ingresa el # filas: "))
        c = int(input("Ingresa el # de columnas: "))
        print("Captura de elementos: ")
        M = []

        #llenando de ceros M lrptm
        for i in range(f):
            M.append([])
            for j in range(c):
                M[i].append(0)

        for i in range(f):
            for j in range(c):
                M[i][j] = int(input(f'Ingresa el elemento {i}, {j}: '))
                M.append(M[i][j])

        print('La matriz es:')
        printMatrix(M)

        return M, f, c
    
    if not A:
        A, no_rows_A, no_cols_A = getMatrix()
        B, no_rows_B, no_cols_B = getMatrix()
    else:
        no_rows_A = len(A)
        no_rows_B = len(B)
        no_cols_A = len(A[0])
        no_cols_B = len(B[0])
        
    A_tensor_B = [] 

    for i in range(no_rows_A*no_rows_B):
        A_tensor_B.append([])
        for j in range(no_cols_A*no_cols_B):
            A_tensor_B[i].append(0)

    for ra in range(no_rows_A): 
        for ca in range(no_cols_A): 
            for rb in range(no_rows_B): 
                for cb in range(no_cols_B):
                    A_tensor_B[ra*no_rows_B + rb][no_cols_B*ca + cb] = A[ra][ca]*B[rb][cb] 
    
    if prints:
        printMatrix(A_tensor_B, 'A \otimes B')
    return A_tensor_B 
```

## Applicar Hadamard

```python
def applyH(v=None, prints=True):
    v_n = vectorNormalizer(v, prints=False)
    
    if (v_n != v) and prints:
        v = v_n
        print('Estado cuntico configurado:')
        printMatrix(v_n, 'v')
    
    value = 1/(2**0.5) 
    H = [ 
        [value, value], 
        [value, -value] 
    ]

    H_v = [[0], [0]]
    
    
    for i in range(2): 
        for j in range(1):
            for k in range(2):
                H_v[i][j] += H[i][k]*v[k][j]
    if prints:
        print('\nResultado:')
        printMatrix(H_v, 'H \cdot v')
    return H_v 
```

## Sacar la forma bra de un ket

```python
def bra(v=None):
    
    if not v:
        v = random_complex_quantum_state()
    
    v[0][0] = v[0][0].conjugate()
    v[1][0] = v[1][0].conjugate()
    
    return TransMatrix(v)
```

```python
from math import sin, cos
def ChangeOfStateB(theta, phi):
    theta = theta/2
    alpha = cos(theta)
    beta = sin(theta)*complex(cos(phi),sin(phi))
    print(f"El estado inicial: cos({theta}/2|0> + e^(i {phi})sin({theta}(2)|1>")
    print(f"Tu estado final es: {alpha} |0> + {beta}|1>")
    return [alpha,beta]
```

```python
qreg =  QuantumRegister(1)
#(verde es el estado inicial y rojo el final)
def initial_circuit():
    creg = ClassicalRegister(1)
    circuit = QuantumCircuit(qreg,creg)
    
    # Change this to begin from different state.
    circuit.h(qreg[0])
    
    return circuit

mycircuit = initial_circuit()
job = execute(mycircuit,Aer.get_backend('statevector_simulator'),optimization_level=0)
current_quantum_state=job.result().get_statevector(mycircuit)
angles1 = state_to_angles(current_quantum_state)

mycircuit = initial_circuit()

# Apply operations for new state here
mycircuit.p(3*pi/2,qreg[0])

job = execute(mycircuit,Aer.get_backend('statevector_simulator'),optimization_level=0)
current_quantum_state2=job.result().get_statevector(mycircuit)

angles2 = state_to_angles(current_quantum_state2)

draw_bloch2(angles1[0], angles1[1], angles2[0], angles2[1])
```

## Vectores Base

```python
def VectoresBase(dim):
    v = []
    
    #llenando de ceros v lrptm
    for i in range(dim):
        v.append([])
        for j in range(dim):
            v[i].append(0)
            
    for i in range(dim):
        v[i][i]=1

    return v
```

## $QFT^{\dagger}$

```python
#%%writefile iqft.py
import cirq
from cirq.circuits import InsertStrategy
from cirq import H, SWAP, CZPowGate

def iqft(n,qubits,circuit):
    
    #Swap the qubits
    for i in range(n//2):
        circuit.append(SWAP(qubits[i],qubits[n-i-1]), strategy = InsertStrategy.NEW)
     
    #For each qubit
    for i in range(n-1,-1,-1):
        #Apply CR_k gates where j is the control and i is the target
        k=n-i #We start with k=n-i
        for j in range(n-1,i,-1):
            #Define and apply CR_k gate
            crk = CZPowGate(exponent = -2/2**(k))
            circuit.append(crk(qubits[j],qubits[i]),strategy = InsertStrategy.NEW)
            k=k-1 #Decrement at each step
            
        #Apply Hadamard to the qubit
        circuit.append(H(qubits[i]),strategy = InsertStrategy.NEW)
```

```python
n=4

#Create circuit
circuit = cirq.Circuit()

#Create n qubits
qubits=cirq.LineQubit.range(n)

#Call qft function
qft(n,qubits,circuit)

#Call iqft function
iqft(n,qubits,circuit)

#Simulating circuit
print('Simulate the circuit:')
s=cirq.Simulator()
results=s.simulate(circuit)
print(results) 

#Unitary matrix representation
print(cirq.unitary(circuit))
```
